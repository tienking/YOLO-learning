```python
from ultralytics import YOLO

# Load a model
model = YOLO("/content/yolov8n.pt")  # build a new model from scratch

# Use the model
model.train(data="/content/drive/MyDrive/YOLO/Gun Detector.v2i.yolov8/data.yaml", epochs=30) 
```

```python
import glob
from os.path import join

imgs = [f for f in glob.glob("/content/drive/MyDrive/YOLO/test/*.jpg")]
results = model(imgs)
```

```python
for result in results:
    boxes = result.boxes.data  # Boxes object for bbox outputs
    for box in boxes:
      print(box)
```

```python
results = model.predict(source='/content/drive/MyDrive/YOLO/Gun Detector.v2i.yolov8/test_image', save=True, save_txt=True)
```
